```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import os


magnitude = 1
angles_amount = 30


f = open("angles.txt", "w")
angles_degrees = range(1, angles_amount + 1)

vectors = []
drags = []
lifts = []

for angle_degrees in angles_degrees:
    angle_radians = np.radians(angle_degrees)
    
    x = magnitude * np.cos(angle_radians)
    y = magnitude * np.sin(angle_radians)
    
    vectors.append((x, y))
    drags.append((x,y))
    lifts.append((-y,x))

    row = "(" + str(x) + " "+  str(y) + " 0);\t(" + str(-y) + " " + str(x) + " 0);\n"
    f.write(row)
    
f.close()


origin = np.zeros((2, len(vectors)))

plt.figure(figsize=(10, 10))
plt.quiver(*origin, [v[0] for v in vectors], [v[1] for v in vectors],
           angles='xy', scale_units='xy', scale=1, color='r', alpha=0.75)


plt.quiver(*origin, [v[0] for v in lifts], [v[1] for v in lifts],
           angles='xy', scale_units='xy', scale=1, color='b', alpha=0.75)


plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: Angles of attack in degrees (you can also use radians)
angles_deg = np.array([1, 5, 10, 15, 20, 25, 30])

# Lift coefficients (C_L) - Replace with your actual data
CL = np.array([55.00336, 123.5035, 173.6059, 128.2145, 144.5061, 160.0227, 173.1441])

# Drag coefficients (C_D) - Replace with your actual data
CD = np.array([4.181834, 5.289954, 11.75121, 38.75057, 58.27483, 80.11029, 104.645])

# Calculate the lift-to-drag ratio (L/D)
L_D_ratio = CL / CD

# Print the lift-to-drag ratios
print("Lift-to-Drag Ratios:")
for angle, ratio in zip(angles_deg, L_D_ratio):
    print(f"Angle of Attack: {angle} -> L/D Ratio: {ratio:.2f}")

# Plotting the lift-to-drag ratio vs. angle of attack
plt.figure(figsize=(10, 6))
plt.plot(angles_deg, L_D_ratio, marker='o', linestyle='-', color='b')
plt.xlabel('Angle of Attack (degrees)')
plt.ylabel('Lift-to-Drag Ratio (L/D)')
plt.title('Lift-to-Drag Ratio vs. Angle of Attack')
plt.grid(True)
plt.show()

# Find the angle of attack with the highest lift-to-drag ratio
max_index = np.argmax(L_D_ratio)
max_angle = angles_deg[max_index]
max_ratio = L_D_ratio[max_index]

print(f"\nThe angle of attack with the highest L/D ratio is {max_angle} with a ratio of {max_ratio:.2f}")
```
