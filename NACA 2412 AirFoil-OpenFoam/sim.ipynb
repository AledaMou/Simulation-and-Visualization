```python
import matplotlib.pyplot as plt
import numpy as np

# Define your x and y coordinates
x = np.array([7.29555304,7.2289569 , 7.1890966, 7.14047749, 7.0831588])
y = np.array([0.25476633, 1.01596364, 1.2676317, 1.51775534, 1.76602984])

# Define the angles in degrees
angles = np.array([30, 45, 60, 120, 150])  # in degrees

# Convert angles to radians for calculation
angles_rad = np.deg2rad(angles)

# Define the length of the arrows (you can adjust this)
arrow_length = 0.5

# Calculate the arrow deltas
dx = arrow_length * np.cos(angles_rad)
dy = arrow_length * np.sin(angles_rad)

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')

# Plot arrows representing angles
for i in range(len(x)):
    plt.arrow(x[i], y[i], dx[i], dy[i], head_width=0.1, head_length=0.1, fc='red', ec='red')

# Add labels and title
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Scatter Plot with Angles')

# Add a grid for better readability
plt.grid(True)

# Set equal scaling for both axes
plt.axis('equal')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Angles of attack
angles = np.array([2, 12, 18, 22, 26, 30])  # Replace with your angles

# Lift and drag coefficients
CL = np.array([80.06101, 130.9383, 146.5151, 165.6822, 182.1582, 207.3228])  # Replace with your lift coefficients
CD = np.array([0.423752, 3.942621, 5.647353, 5.381763, 4.911845, 3.262038])  # Replace with your drag coefficients

# Calculate lift-to-drag ratio
L_D_ratio = CL / CD

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(angles, L_D_ratio, marker='o', linestyle='-', color='b')
plt.xlabel('Angle of Attack (degrees)')
plt.ylabel('Lift-to-Drag Ratio (L/D)')
plt.title('Lift-to-Drag Ratio vs. Angle of Attack')
plt.grid(True)
plt.show()

# Identify the maximum L/D ratio
max_index = np.argmax(L_D_ratio)
print(f'The angle of attack with the highest L/D ratio is {angles[max_index]} degrees')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import os


magnitude = 7.3
angles_amount = 30


f = open("angles.txt", "w")
angles_degrees = range(1, angles_amount + 1)

vectors = []

for angle_degrees in angles_degrees:
    angle_radians = np.radians(angle_degrees)
    
    x = magnitude * np.cos(angle_radians)
    y = magnitude * np.sin(angle_radians)
    
    vectors.append((x, y))
    row = "internalField   uniform (" + str(x) + " "+  str(y) + " 0);\n"
    f.write(row)
    
f.close()
origin = np.zeros((2, len(vectors)))

plt.figure(figsize=(10, 10))
plt.quiver(*origin, [v[0] for v in vectors], [v[1] for v in vectors],
           angles='xy', scale_units='xy', scale=1, color='b', alpha=0.75)


plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
